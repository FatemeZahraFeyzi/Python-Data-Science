## distplot = distribution of the univariant(one variable) set of observation

```python
sns.distplot(tips['total_bill'])
```

```python
## it will remove the kde(kernel density extimation) diagram
sns.distplot(tips['total_bill'],kde=False)
```

```python
#bins= it shows the number of parts that x axis will divided to
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## jopint plot= it allows you to combine two different distplot(two variables)

```python
## use x and y to define wich variables(columns,features) you are going to compare
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
# kind parameter allows you to set the type of inside diagram
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

## pairplot will show pairwise relationship across an entire dataframe or at least for numereical columns

```python
#hue paremeter is for passing the name of categorical columns and seperate data base on it
sns.pairplot(tips,hue='sex')
```

```python
#palette alows you to color your plot
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
# despite of histogram this plot is horizental ;)
sns.rugplot(tips['total_bill'])
```

```python
# kde (kernel density estimation)
sns.kdeplot(tips['total_bill'])
```

## categorical plots

```python
import seaborn as sns
%matplotlib inline
tips= sns.load_dataset('tips')
tips.head()
```

```python
# it shows the average or mean a non-categoricalk column base of categorical column
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
## by deafault the estimator is mean but we can change it
## this exacly doing group by mechanism
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#this is the barplot but the estimator is the number of each category
sns.countplot(x='sex',data=tips)
```

## boxplot is used to show distribution on categorical variable
box or whisker plot


```python
# for each category of categorical variable it shows the box plot
sns.boxenplot(x='day',y='total_bill',data=tips)
```

```python
# add another categorical variable
sns.boxenplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxenplot(x='day',y='total_bill',data=tips)
```

```python
#it will show the distribution of data across some sort of category
# this plot unlike the box plot allows us to actually plot all the components that 
## correspond to actual data points and it is essentially showing the kernel density eatimation of underlying distribution
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## stript plot

```python
#it will draw a scatter plot when one variable is categorical
sns.stripplot(x='day',y='total_bill',data=tips)
# this is an issue with scatter plot that you dont khow how many points are stack on top of each other
```

```python
# to fix that above issue
# jitter will add a little bit of random noise to seperate some of stacked points
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

## swarm plot

```python
#its simmilar to striptplot except that points are adjusted and dont overlap
sns.swarmplot(x='day',y='total_bill',data=tips)
# not to use this plot for a large data set
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# there is a kind argument is going to describe what is going on with factor plot itself
# it will define kind of chart
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

## matrix plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
flights=sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

### in order to heatmap do well the data must be in matrix for(both columns and rows must be variables)

```python
# first way to get data in matrix form
tc=tips.corr()
```

```python
sns.heatmap(tc)
```

```python
#annot=true will show the actual value of each cell
#cmap will define a color map
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

# we can create a matrix form by pivote_table

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp)
```

```python
# magma create colors from dark to light
sns.heatmap(fp,cmap='magma')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=2)
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidth=2)
```

```python
sns.heatmap(fp)
```

## clustermap = it uses the hierarichal clustering to produce a clustered version of this heatmap

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
# standard scale will normalize the scale of value
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```
